class ContactBook:
- Create Contact
- Delete Contact
- Update Contact
- Show All Contacts
- Search Contact
- Exit

```python
import json

class ContactBook:
    
    def __init__(self, file_path):
        self.file_path = file_path
        self.contacts = self.load_contacts()  # Load contacts when initializing

    # Load existing contacts from JSON file
    def load_contacts(self):
        try:
            with open(self.file_path, 'r') as file:
                return json.load(file)
        except FileNotFoundError:
            return {}  # Return an empty dictionary if the file doesn't exist

    # Save updated contacts to JSON file
    def save_contacts(self):
        with open(self.file_path, 'w') as file:
            json.dump(self.contacts, file, indent=4)

    # Add a new contact
    def add_contact(self, name, last_name, phone_number, ID, email=None, location=None):
        if ID in self.contacts:
            print(f"Contact with ID {ID} already exists.")
            return
        
        # Add contact to the in-memory dictionary
        self.contacts[ID] = {
            'name': name,
            'last_name': last_name,
            'phone_number': phone_number,
            'email': email,
            'location': location
        }
        
        # Save to file after adding
        self.save_contacts()
        print(f"Contact {ID} added successfully.")

    # View all contacts
    def view_contacts(self):
        if not self.contacts:
            print("No contacts available.")
            return
        
        for ID, info in self.contacts.items():
            print(f"ID: {ID}")
            print(f"Name: {info['name']}")
            print(f"Last Name: {info['last_name']}")
            print(f"Phone Number: {info['phone_number']}")
            print(f"Email: {info['email']}")
            print(f"Location: {info['location']}")
            print("----------------------------------")


# Example usage
file_path = 'contact_book_with_ids.json'
contact_book = ContactBook(file_path)

# View existing contacts
contact_book.view_contacts()

# Add a new contact
contact_book.add_contact(
    name="Aghigh",
    last_name="Dadashzadeh",
    phone_number="09111111111",
    ID="2640253190",
    email="aghighaghigh@gmail.com",
    location="Rasht"
)

# View contacts after adding a new one
contact_book.view_contacts()
```
